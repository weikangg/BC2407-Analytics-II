#### Total columns from Vehicle.csv = 22 variables + 2 (DRIVERRF & DRDISTRACT) - 1 (pveh_sev)  - 2 (only used for merging) = 21.

````python
filtered_vehicle_df = vehicle_df.iloc[ : , [2,3,5,21,25,60,64,80,116,118,130,142,144,146,152,154,156,162,164,175,187]] 

filtered_vehicle_df
````

````python
distract_df = pd.read_csv("../FARS2020NationalCSV/Distract.csv", encoding='latin-1')

filtered_distract_df = distract_df.iloc[ : , [2,3,4]] 

filtered_distract_df
````

````python
new_filtered_vehicle_df= filtered_vehicle_df.merge(filtered_distract_df, on=['ST_CASE', 'VEH_NO'], how='inner')

new_filtered_vehicle_df
````

````python
## After merging, we had some duplicate rows, as we see that the total number of rows is 54291, which is according to the number of rows in distract.df. hence, we drop the duplicate rows.

new_filtered_vehicle_df.drop_duplicates(subset=['ST_CASE','VEH_NO'], inplace=True)

new_filtered_vehicle_df
````

````python
driverRF_df = pd.read_csv("../FARS2020NationalCSV/driverrf.csv", encoding='latin-1')

filtered_driverRF_df = driverRF_df.iloc[ : , [2,3,4]] 

filtered_driverRF_df
````

````python
new_filtered_vehicle_df= new_filtered_vehicle_df.merge(filtered_driverRF_df, on=['ST_CASE', 'VEH_NO'], how='inner')

new_filtered_vehicle_df
````

````python
## We do the same thing, to remove all the duplicates so that the number of rows will be 54272 instead of 59982.

new_filtered_vehicle_df.drop_duplicates(subset=['ST_CASE','VEH_NO'], inplace=True)

new_filtered_vehicle_df
````

````python
accident_df = pd.read_csv("../FARS2020NationalCSV/accident.csv", encoding='latin-1')

filtered_accident_df = accident_df.iloc[ : , [2,14,16,18,19,21,31,45,51,53,57,59,61,82]] 

filtered_accident_df
````

### Merging Person.csv and Vehicle.csv on ST_CASE and VEH_NO

````python
personAndVehicle_df=pd.merge(filtered_person_df,new_filtered_vehicle_df, on=['ST_CASE', 'VEH_NO'], how='inner')
personAndVehicle_df

````

````python
personAndVehicle_df.drop_duplicates(subset=['ST_CASE','VEH_NO'], inplace=True)

personAndVehicle_df
````

````python
final_df = pd.merge(personAndVehicle_df,filtered_accident_df, on=['ST_CASE'], how='inner')
final_df
````

````python
final_df.drop_duplicates(subset=['ST_CASE'], inplace=True)

final_df['RUR_URB'].value_counts()
````
