```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv(r"C:\Users\Loy\Documents\GitHub\BC2407-Analytics-II-Project\Dataset\trafficAccident2020_cleaned.csv")
df
```

```python
df.describe()
```

```python
plt.hist(df['Fatality_Rate'],bins = 50, edgecolor='black')
plt.title('Distribution of Fatality_Rate')
plt.show()
```

```python
grouped_data = df.groupby("Travel_Speed").mean()["Fatality_Rate"]
plt.plot(grouped_data.index, grouped_data.values)
plt.title("Fatality Rate against Travel Speed")
plt.xlabel("Travel Speed") # Set y-axis limits from 0 to 10
```

```python
grouped_data = df.groupby("Day_Of_Week").mean()["Num_Of_Occupants"]
plt.plot(grouped_data.index, grouped_data.values,marker='o')
plt.title("Mean Num of Occupants Per Day of Week")
plt.xlabel("Day of Week")
plt.ylim(bottom=1, top=2)  # Set y-axis limits from 0 to 10
for x, y in zip(grouped_data.index, grouped_data.values):
    plt.text(x, y, str(round(y, 2)), ha='center', va='bottom')
```

```python
plt.hist(df['Fatality_Rate'], bins=50, edgecolor='black')
plt.title('Distribution of Fatality_Rate')
plt.show()
```

```python
grouped_data = df.groupby("Num_Of_Lanes").mean()["Fatality_Rate"]
plt.plot(grouped_data.index, grouped_data.values,marker='o')
plt.title("Mean Fatality Rate per Num Of Lanes")
plt.xlabel("Fatality Rate")
plt.ylim(bottom=1, top=2)  # Set y-axis limits from 0 to 10
for x, y in zip(grouped_data.index, grouped_data.values):
    plt.text(x, y, str(round(y, 2)), ha='center', va='bottom')
```
