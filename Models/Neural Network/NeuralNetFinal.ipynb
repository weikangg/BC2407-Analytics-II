```python
import tensorflow as tf
```

```python
# using feature_columns in tensorflow to format data for NN
feature_columns = []
numeric_headers = list(X.select_dtypes(include=['int64','float64']).columns)
for header in numeric_headers:
  feature_columns.append(tf.feature_column.numeric_column(header))
feature_columns
```

```python
factors = list(X.select_dtypes(include=['category']).columns)
for i in factors:
    feature = tf.feature_column.categorical_column_with_vocabulary_list(i, X[i].unique())
    feature_emb = tf.feature_column.embedding_column(feature, dimension=32) # Start w 8 dimensions for computation speed
    feature_columns.append(feature_emb)
print(feature_columns)
```

```python
df['Death']=df['Death'].replace('No',0)
df['Death']= df['Death'].replace('Yes',1)
Y=df['Death']
```

```python
df= pd.concat([X,Y],axis=1)
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(df, test_size=0.2)
train, val  = train_test_split(train, test_size=0.2)
# Converting Dataframes to Dataset.
def df_to_dataset(df, shuffle=True, batch_size=32):
  df = df.copy()
  labels = df.pop('Death')
  ds = tf.data.Dataset.from_tensor_slices((dict(df) , labels))
  if shuffle:
    ds.shuffle(buffer_size=len(df))
  ds = ds.batch(batch_size)
  return ds
#Creating Dataset for training, validation and testing datatframes.
batch_size=64
train_ds = df_to_dataset(train, shuffle=True , batch_size=batch_size)
val_ds   = df_to_dataset(val  , shuffle=False, batch_size=batch_size)
test_ds  = df_to_dataset(test , shuffle=False, batch_size=batch_size)
```

```python
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense,Dropout
```

```python
feature_layer = tf.keras.layers.DenseFeatures(feature_columns)
```

```python
model = tf.keras.Sequential([
        feature_layer,
        tf.keras.layers.Dense(128, activation='relu'),
        tf.keras.layers.Dense(128, activation='relu'),
        tf.keras.layers.Dense(1)
])
model.compile(loss="binary_crossentropy",optimizer="adam",metrics=["accuracy"]) #loss is during epoch #metrics measures cost at the end of epoch
#dont use accuracy when false negative and false positive is not symmetrical
#optimzer is adaptive learning rate , learning rate depends via momentum(adam optimzer formula).
```

```python
h=model.fit(train_ds, validation_data=val_ds, epochs=100)
```

```python
pred=model.predict(test_ds)
pred = pred>0.5
pred = pred.astype('int32')
pred
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(test['Death'],pred)
print("accuracy is : ",(cm[0,0]+cm[1,1])/sum(sum(cm)))
```

```python
print(cm)
```

```python
cm[1,0]
```

```python
#Accuracy = (TP+TN)/(TP+FP+FN+TN)
accuracy = (cm[0,0]+cm[1,1])/sum(sum(cm))
print("accuracy is : ",accuracy)
#Precision = (TP)/(TP+FP)
precision = cm[0,0]/sum(cm[0])
print("precision is : ",precision)
#Recall = (TP)/(TP+FN)
recall = cm[0,0]/(cm[0,0]+cm[1,0])
print("recall is : ",recall)
#False Positive Rate = FP/ (FP+TN)
FPR =  cm[0,1]/(cm[0,1]+cm[1,1])
print("False Positive Rate  is : ",FPR)
#False Negative Rate = FN / (FN+TP)
FNR =  cm[1,0]/(cm[1,0]+cm[0,0])
print("False Negative Rate is : ",FNR)
```

```python
#F1  = TP/(TP +0.5 (FP + FN)) = 2*((precision*recall)/(precision+recall))
F1 = 2*((precision*recall)/(precision+recall))
print("F1 is : ",F1)

#F2-measure puts more attention on minimizing false negatives than minimizing false positives.
#F2-Measure = (5 * Precision * Recall) / (4 * Precision + Recall)
F2 = (5 * precision * recall)/(4* precision + recall)
print("F2 is : ",F2)
```
