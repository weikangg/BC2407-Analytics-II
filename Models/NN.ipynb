```python
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(df, test_size=0.2)
train, val  = train_test_split(train, test_size=0.2)
# Converting Dataframes to Dataset.
def df_to_dataset(df, shuffle=True, batch_size=32):
  df = df.copy()
  labels = df.pop('Fatality_Rate')
  ds = tf.data.Dataset.from_tensor_slices((dict(df) , labels))
  if shuffle:
    ds.shuffle(buffer_size=len(df))
  ds = ds.batch(batch_size)
  return ds
#Creating Dataset for training, validation and testing datatframes.
batch_size=64
train_ds = df_to_dataset(train, shuffle=True , batch_size=batch_size)
val_ds   = df_to_dataset(val  , shuffle=False, batch_size=batch_size)
test_ds  = df_to_dataset(test , shuffle=False, batch_size=batch_size)
```

```python
feature_layer = tf.keras.layers.DenseFeatures(feature_columns)
```

```python
from tensorflow import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense,Dropout
```

```python
model = Sequential()
```

```python
model = tf.keras.Sequential([
        feature_layer,
        tf.keras.layers.Dense(128, activation='relu'),
        tf.keras.layers.Dense(128, activation='relu'),
        tf.keras.layers.Dense(1)
])
model.compile(loss="mse",optimizer="adam",metrics=["mse"]) #loss is during epoch #metrics measures cost at the end of epoch
#dont use accuracy when false negative and false positive is not symmetrical
#optimzer is adaptive learning rate , learning rate depends via momentum(adam optimzer formula).
```

```python
model.fit(train_ds, validation_data=val_ds, epochs=100)
```

```python
loss, accuracy = model.evaluate(test_ds)
loss
accuracy
```

```python
from sklearn.metrics import mean_squared_error
pred=model.predict(test_ds)
print("rmse is ",mean_squared_error(test['Fatality_Rate'],pred)**0.5)
```
