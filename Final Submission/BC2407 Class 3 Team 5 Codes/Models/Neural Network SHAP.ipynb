```python
Con_Var = list(X.select_dtypes(include=['float64']))
Con_Var.append('Age')
Con_Var
```

```python
#Normalizing Continuous Variables
from scipy import stats 

def zcolumn(column):
    global X
    X[column] = stats.zscore(X[column])

for i in Con_Var:
    zcolumn(i)

X
```

```python
X = pd.get_dummies(X)
```

```python
import tensorflow as tf
```

```python
df['Death']=df['Death'].replace('No',0)
df['Death']= df['Death'].replace('Yes',1)
Y=df['Death']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test  = train_test_split(X,Y, test_size=0.2)
```

```python
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense,Dropout
```

```python
model = tf.keras.Sequential([
        tf.keras.layers.Dense(128, activation='relu'),
        tf.keras.layers.Dense(128, activation='relu'),
        tf.keras.layers.Dense(1)
])
model.compile(loss="binary_crossentropy",optimizer="adam",metrics=["accuracy"]) #loss is during epoch #metrics measures cost at the end of epoch
#dont use accuracy when false negative and false positive is not symmetrical
#optimzer is adaptive learning rate , learning rate depends via momentum(adam optimzer formula).
```

```python
h=model.fit(X_train,Y_train,batch_size=16,epochs=100)
```

```python
pred=model.predict(X_test)
pred = pred>0.5
pred = pred.astype('int32')
pred
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(Y_test,pred)
print("accuracy is : ",(cm[0,0]+cm[1,1])/sum(sum(cm)))
```

```python
print(cm)
```

```python
cm[1,0]
```

```python
#Accuracy = (TP+TN)/(TP+FP+FN+TN)
accuracy = (cm[0,0]+cm[1,1])/sum(sum(cm))
print("accuracy is : ",accuracy)
#Precision = (TP)/(TP+FP)
precision = cm[0,0]/sum(cm[0])
print("precision is : ",precision)
#Recall = (TP)/(TP+FN)
recall = cm[0,0]/(cm[0,0]+cm[1,0])
print("recall is : ",recall)
```

```python
#F2-measure puts more attention on minimizing false negatives than minimizing false positives.
#F2-Measure = (5 * Precision * Recall) / (4 * Precision + Recall)
F2 = (5 * precision * recall)/(4* precision + recall)
print("F2 is : ",F2)
```
